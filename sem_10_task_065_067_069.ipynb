Задача №67. Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)

```python
penguins
```

```python
def type_bill(x):
  if x < 35:
    return 'low'
  elif x > 35 and x < 42:
    return 'middle'
  elif x > 42:
    return 'high'
penguins['type_bill'] = penguins.bill_length_mm.apply(type_bill)   # название нового столбца = столбец, из которого мы его создаем
```

```python
penguins
```

```python
sns.scatterplot(x = penguins.bill_length_mm, y = penguins.bill_depth_mm, hue=penguins.type_bill)
```
